```python
role = '''
You are a behavioral social scientist interviewing Florida residents to gather responses to hurricane risk.
'''
prompt = '''
A Category 5 hurricane is forecasted to hit your county within the next 18 months. Consider this imminent, catastrophic threat alongside the growing, long-term risks and intensities of hurricanes to your county where you live in Florida, especially due to the effects of climate change. Would you consider migrating out of your county (moving to a different county or state) given these risks? Answer with a percent probability from 0% to 100% where 0% represents a 0% chance of moving and 100% represents a 100% chance of moving within the next few years. List the top 2â€“3 reasons that justify the predicted likelihood and decision to migrate. The reasoning must logically match the probability response. If 0%-50%, then will likely NOT consider moving: the persona has an overwhelming, immutable attachment to their location, employment, or social network; they view any risk as manageable or irrelevant compared to the social/financial cost of leaving; the attachment dominates the decision. If 50%-100%, then will strongly consider moving: the persona perceives the risk as a threat that their current resources (financial, preparedness, etc.) cannot withstand, regardless of social ties or attachment; the risk dominates the decision.
Consider the following constraints: You must balance risk perception, socio-economic constraints, place attachment, institutional trust, prior hurricane experience, and more. Assume that these outputs will be used in simulation studies and policy experiments. Avoid fabricating specific addresses or employers. Keep language non-clinical and non-pathologizing. Do not reference statistical terms (e.g., regression, coefficients) in the reasons. Check that reasons reference different dimensions (e.g., risk perception, resources, affordability, social ties) rather than repeating the same factor. If any prediction seems implausible given basic social-science intuition, silently adjust willingness, decision, or reasons to restore coherence.
The decision will vary based on the persona details. Each persona will have the following information: COUNTY (name of the county in Florida), AGE (age of the individual in years), INCOME (approximate annual income in USD), HOME_VALUE (estimated market value of the individual's home in USD), EDUCATION (highest level of education attained), PREPAREDNESS (self-assessed level of hurricane preparedness), RISK_RESPONSE (perceived likelihood of hurricane affecting household), KIDS (number of children in the household), EMPLOYMENT (current employment status), STATUS (current marital status), OWNERSHIP (current home ownership status), CONSTRUCTION_YEAR (year the house was built), FLOOD_ZONE (flood zone classification of the county), STORM_COUNT (number of historical storms in the area), WIND_AVG (average wind speed in the area in mph), FLOOD_DEPTH (average flood depth in the area in feet), ELEVATION_DIFF (difference between lowest floor elevation and base flood elevation where positive = above BFE, negative = below BFE).
We ran 3 models (GLMER, Random Forest, XGBoost) to derive the feature importances of each variable to see how it impacts migration decisions. The overall top variables influencing migration decisions were, in order from most important: age, home ownership, number of children, home value, education, and marital status. Consider these variables first when deciding likelihood to migrate. The weights for some of these variables are: AGE=0.037, HOME_VALUE=0.029, EDUCATION=0.022, KIDS=0.021, OWNERSHIP=0.407.
More information on flood zone classifications: Flood hazard areas identified on the Flood Insurance Rate Map are identified as a Special Flood Hazard Area (SFHA). SFHA are defined as the area that will be inundated by the flood event having a 1-percent chance of being equaled or exceeded in any given year. The 1-percent annual chance flood is also referred to as the base flood or 100-year flood. SFHAs are labeled as Zone A, Zone AO, Zone AH, Zones A1-A30, Zone AE, Zone A99, Zone AR, Zone AR/AE, Zone AR/AO, Zone AR/A1-A30, Zone AR/A, Zone V, Zone VE, and Zones V1-V30. Moderate flood hazard areas, labeled Zone B or Zone X (shaded) are also shown on the FIRM, and are the areas between the limits of the base flood and the 0.2-percent-annual-chance (or 500-year) flood. The areas of minimal flood hazard, which are the areas outside the SFHA and higher than the elevation of the 0.2-percent-annual-chance flood, are labeled Zone C or Zone X (unshaded).
For the distinct flood zone groupings present in the persona data (i.e., A, V, X), consider the following insurance information to aid your decision to migrate. The total coverage represents the total amount of building and contents coverage in USD. Total coverage: all A zones (high risk) = 190,546,647,000; all V zones (high risk) = 14,541,779,000; all X zones (low risk) = 135,669,128,000. Include in your decision if having/not having insurance on home/belongings would affect hurricane migration decisions.
'''
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter
import seaborn as sns
```

```python
len(probs)
```

```python
sns.reset_defaults()
```

```python
sns.displot(probs, kde=True, rug=True, rug_kws={"alpha": 0.5, "linewidth": 8, "height": 0.1})
plt.title('Histogram of Probabilities')
plt.xlabel('Probability of Migration')
plt.show()
```

```python
probs = np.array(probs)

probs_low = probs[probs < 0.5]
probs_high = probs[probs >= 0.5]

fig, ax = plt.subplots(figsize=(8, 5))

sns.histplot(
    x=probs_low,
    ax=ax,
    color="sandybrown",
    label="p < 0.5: NO",
    kde=False
)

sns.histplot(
    x=probs_high,
    ax=ax,
    bins=np.linspace(0, 1, 30),
    color="skyblue",
    label="p >= 0.5: YES",
    kde=False,
    line_kws={'linewidth': 2}
)

sns.rugplot(
    x=probs_low,
    ax=ax,
    color="darkorange",
    alpha=0.5,
    linewidth=8,
    height=0.1
)

sns.rugplot(
    x=probs_high,
    ax=ax,
    color="dodgerblue",
    alpha=0.5,
    linewidth=8,
    height=0.1
)

ax.set_xticks(np.arange(0.0, 1.1, 0.1))
ax.xaxis.set_major_formatter(StrMethodFormatter('{x:.2f}'))

ax.set_title('Histogram of Probabilities')
ax.set_xlabel('Probability of Migration')
ax.set_ylabel('Count')

# ax.set_xlim(0, 1)
ax.axvline(0.5, color='black', linestyle='--', linewidth=1)

plt.legend(title='Migration Decision')
plt.show()
```
